## problem 1 (a)

Select the entry (or exit) data of 1 specific month, and plot the time series of 4 stations (overlapped on the same line chart).

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 

df_january = pd.read_csv('202501_en.csv')
df_february = pd.read_csv('202502_en.csv')
df_march = pd.read_csv('202503_en.csv')

#dates
dates_january = column_1_data = df_january[df_january.columns[0]].to_numpy()
dates_february = column_1_data = df_february[df_february.columns[0]].to_numpy()
dates_march = column_1_data = df_march[df_march.columns[0]].to_numpy()

#station names january,february,march since names wont change
station_1 = df_january.columns[1]
station_2 = df_january.columns[2]
station_3 = df_january.columns[3]
station_4 = df_january.columns[4]


# Convert data to integers after removing commas for january
column_1_data_int_january = df_january[df_january.columns[1]].str.replace(',', '').astype(int) #station 1
column_2_data_int_january = df_january[df_january.columns[2]].str.replace(',', '').astype(int) #station 2
column_3_data_int_january = df_january[df_january.columns[3]].str.replace(',', '').astype(int) #station 3
column_4_data_int__january = df_january[df_january.columns[4]].str.replace(',', '').astype(int) #station 4

# Convert data to integers after removing commas for february
column_1_data_int_february = df_february[df_february.columns[1]].str.replace(',', '').astype(int)  #station 1
column_2_data_int_february = df_february[df_february.columns[2]].str.replace(',', '').astype(int)   #station 2
column_3_data_int_february = df_february[df_february.columns[3]].str.replace(',', '').astype(int)   #station 3
column_4_data_int_february = df_february[df_february.columns[4]].str.replace(',', '').astype(int)   #station 4

# Convert data to integers after removing commas for march
column_1_data_int_march = df_march[df_march.columns[1]].str.replace(',', '').astype(int)    #station 1
column_2_data_int_march = df_march[df_march.columns[2]].str.replace(',', '').astype(int)    #station 2
column_3_data_int_march = df_march[df_march.columns[3]].str.replace(',', '').astype(int)    #station 3
column_4_data_int_march = df_march[df_march.columns[4]].str.replace(',', '').astype(int)    #station 4

```

```python
first_value = dates_january[0]
def plot_station_timeseries(start_date, values, station_name, new_fig=False):
    data = {
        'Date': pd.date_range(start=start_date, periods=len(values), freq='D'),
        'Value': values
    }

    df = pd.DataFrame(data)
    df.set_index('Date', inplace=True)

    # Only create a new figure if requested
    if new_fig:
        plt.figure(figsize=(16, 6))

    plt.plot(df.index, df['Value'], marker='o', label=station_name)

    plt.xlabel('Date')
    plt.ylabel('Ridership counts')
    plt.grid(True)

    # Show ALL dates
    # plt.xticks(df.index, df.index.strftime("%Y-%m-%d"), rotation=45, ha='right')

    plt.legend()

# Example usage:
plot_station_timeseries(first_value, column_1_data_int_january, station_1, new_fig=True)
plot_station_timeseries(first_value, column_2_data_int_january, station_2)
plot_station_timeseries(first_value, column_3_data_int_january, station_3)
plot_station_timeseries(first_value, column_4_data_int__january, station_4)

# Finalize once
plt.title("Time Series - All Stations (January)")
plt.tight_layout()
plt.show()
```

## Problem 1 (b)

Choose 1 station and collect its data for 3 consecutive months. Then, plot the box plot and histogram for this station during the time period.


```python
data1 = column_1_data_int_january
data2 = column_1_data_int_february
data3 = column_1_data_int_march

data = [data1, data2, data3]

plt.boxplot(data, labels=["January", "February", "March"])
plt.title(station_1)
plt.ylabel("Ridership counts")  
plt.show()
```

```python
#create a histogram for the three months
plt.figure(figsize=(10, 6))
plt.hist(data, bins=15, label=["January", "February", "March"], alpha=0.7)
plt.title(f"Histogram of Ridership Counts - {station_1}")
plt.xlabel("Ridership counts")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Problem 2 

Using Python (pandas and numpy), apply a moving average method with a window size between 3 and 7 days to the entry or exit ridership data of 1 specific station across 3 consecutive months. then plot both the original data and the data after applying the moving average on the same line chart.


```python
window_size = 7

# Convert array of integers to pandas series
numbers_series = pd.Series(column_1_data_int_3months)

# Get the window of series
# of observations of specified window size
windows = numbers_series.rolling(window_size)

# Create a series of moving
# averages of each window
moving_averages = windows.mean()

# Convert pandas series back to list
moving_averages_list = moving_averages.tolist()

# Remove null entries from the list
final_list = moving_averages_list[window_size - 1:]

print(final_list)

# column_1_data_int_3months = pd.concat([column_1_data_int_january, column_1_data_int_february, column_1_data_int_march], ignore_index=True)

plot_station_timeseries(first_value, column_1_data_int_3months, station_1, new_fig=True)
plot_station_timeseries(first_value, final_list, "moving average")

# Finalize once
plt.title("Time Series - Moving Average")
plt.tight_layout()
plt.show()
```
